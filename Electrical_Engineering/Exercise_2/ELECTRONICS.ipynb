# **Άσκηση 3**

Κατά τη διάρκεια ενός πειράματος στο εργαστήριο υλοποιήθηκε το κύκλωμα (σχήμα 1) για την εξαγωγή της χαρακτηριστικής καμπύλης εξόδου του διπολικού τρανζίστορ επαφής.

![Διάγραμμα](image%20(6).png)

Σχήμα 1: Κύκλωμα πόλωσης τρανζίστορ κοινού εκπομπού με πόλωση βάσης, για την εξαγωγή της χαρακτηριστικής καμπύλης εξόδου

Στο παραπάνω κύκλωμα πόλωσης βάσης, τα μεγέθη για τις αντιστάσεις και τις τάσεις που χρησιμοποιήθηκαν κατά τη διάρκεια του εργαστηρίου είναι R1 = 1ΜΩ, R2 = 1ΚΩ και VBB = 5V. Με δεδομένο ότι Vcc = 10V και ο συντελεστής ενίσχυσης (κέρδος) του τρανζίστορ είναι β = 500, να γίνει η DC ανάλυση του κυκλώματος:

Α) **Υπολογίζοντας θεωρητικά το σημείο λειτουργίας Q**

Β) **Σχεδιάζοντας τη γραμμή φορτίου** επάνω στη χαρακτηριστική καμπύλη. Σε ποια περιοχή λειτουργίας βρίσκεται το τρανζίστορ (δώστε απάντηση με βάση τη θέση του σημείου Q στη χαρακτηριστική καμπύλη εξόδου);

````python
# Given values
R1 = 1e6  # 1 MΩ
R2 = 1e3  # 1 kΩ
VBB = 5  # 5 V
Vcc = 10  # 10 V
beta = 500  # Current gain of the transistor

# Base current calculation
IB = (VBB - 0.7) / R1  # Subtracting 0.7V for the base-emitter voltage drop

# Collector current calculation
IC = beta * IB

# Voltage across R2
VR2 = IC * R2

# Collector-emitter voltage
VCE = Vcc - VR2

print(f"IB is:{IB}\nIC is: {IC}\nVCE is: {VCE}")
````

````python
import numpy as np
import plotly.graph_objects as go

# -----------------------------------
# Παράμετροι τρανζίστορ και κυκλώματος
# -----------------------------------
beta = 500        # Συντελεστής ενίσχυσης
R2 = 1e3          # 1 kΩ
VCC = 10.0        # 10 V
VCE_sat = 0.2     # Τάση κορεσμού (προσεγγιστικά)
# Ρεύματα βάσης (διάφορες τιμές για τις καμπύλες)
Ib_values = [5e-6, 10e-6, 15e-6, 20e-6, 25e-6]

# -----------------------------------
# Προετοιμασία "άξονα" VCE
# -----------------------------------
VCE_points = np.linspace(0, 10, 200)  # από 0 έως 10 V

# Δημιουργία γραφήματος
fig = go.Figure()

# -----------------------------------
# Σχεδίαση χαρακτηριστικών εξόδου
# -----------------------------------
for Ib in Ib_values:
    Ic_curve = []
    for vce in VCE_points:
        if vce <= VCE_sat:
            Ic_curve.append(beta * Ib)
        else:
            Ic_curve.append(beta * Ib)
    fig.add_trace(go.Scatter(
        x=VCE_points, 
        y=Ic_curve, 
        mode='lines', 
        name=f'Ib = {Ib*1e6:.0f} μA',
        hoverinfo='text',
        hovertext=[f'VCE: {vce:.2f} V<br>IC: {ic:.6f} A' for vce, ic in zip(VCE_points, Ic_curve)]
    ))

# -----------------------------------
# Σχεδίαση ευθείας φορτίου (Load Line)
# -----------------------------------
Ic_load_line = (VCC - VCE_points) / R2
fig.add_trace(go.Scatter(
    x=VCE_points, 
    y=Ic_load_line, 
    mode='lines', 
    name='Load Line',
    line=dict(dash='dash'),
    hoverinfo='text',
    hovertext=[f'VCE: {vce:.2f} V<br>IC: {ic:.6f} A' for vce, ic in zip(VCE_points, Ic_load_line)]
))

# -----------------------------------
# Υπολογισμός σημείου λειτουργίας (Q-point)
# -----------------------------------
R1 = 1e6
VBB = 5.0
VBE = 0.7
Ib = (VBB - VBE) / R1  # ρεύμα βάσης
IcQ = beta * Ib        # ρεύμα συλλέκτη
VCEQ = VCC - IcQ * R2  # τάση συλλέκτη-εκπομπού

# Σχεδιάζουμε το σημείο Q
fig.add_trace(go.Scatter(
    x=[VCEQ], 
    y=[IcQ], 
    mode='markers+text', 
    name='Q-point',
    text=[f'Q\n(VCE={VCEQ:.2f} V,\nIC={IcQ*1e3:.2f} mA)'],
    textposition='top right',
    marker=dict(color='black', size=10)
))

# Προσθήκη τίτλων και ετικετών
fig.update_layout(
    title="Χαρακτηριστικές εξόδου τρανζίστορ με την ευθεία φορτίου",
    xaxis_title="VCE (V)",
    yaxis_title="IC (A)",
    xaxis=dict(showline=True, zeroline=True),
    yaxis=dict(showline=True, zeroline=True),
    legend=dict(x=1.05, y=1)  # Move the legend outside the main graph area
)

# Εμφάνιση γραφήματος
fig.show()
````

# **Δεδομένα Κυκλώματος**
Ακολουθούν τα δεδομένα που δόθηκαν στην εκφώνηση και χρησιμοποιήθηκαν για την επίλυση:

| Παράμετρος      | Τιμή                 |
|------------------|----------------------|
| $$ R_1 $$       | $$ 1 \, M\Omega $$  |
| $$ R_2 $$       | $$ 1 \, k\Omega $$  |
| $$ V_{BB} $$    | $$ 5 \, V $$         |
| $$ V_{CC} $$    | $$ 10 \, V $$        |
| $$ V_{BE} $$    | $$ 0.7 \, V $$       |
| Συντελεστής β   | $$ 500 $$            |

---

#### **Υπολογισμοί Βήμα προς Βήμα**

1. **Υπολογισμός Ρεύματος Βάσης ($ I_B $):**

   Το ρεύμα βάσης υπολογίζεται από τη σχέση:
   $
   I_B = \frac{V_{BB} - V_{BE}}{R_1}
   $

   Αντικαθιστώντας τις τιμές:
   $
   I_B = \frac{5 - 0.7}{1 \cdot 10^6} = 4.3 \, \mu A
   $

2. **Υπολογισμός Ρεύματος Συλλέκτη ($ I_C $):**

   Το ρεύμα συλλέκτη σχετίζεται με το ρεύμα βάσης μέσω του συντελεστή ενίσχυσης ($ β $):
   $
   I_C = β \cdot I_B
   $

   Αντικαθιστώντας τις τιμές:
   $
   I_C = 500 \cdot (4.3 \cdot 10^{-6}) = 2.15 \, mA
   $

3. **Υπολογισμός Τάσης Συλλέκτη-Εκπομπού ($ V_{CE} $):**

   Η τάση συλλέκτη-εκπομπού υπολογίζεται από τη σχέση:
   $
   V_{CE} = V_{CC} - I_C R_2
   $

   Αντικαθιστώντας τις τιμές:
   $
   V_{CE} = 10 - (2.15 \cdot 10^{-3}) \cdot (1 \cdot 10^3) = 7.85 \, V
   $

4. **Γραμμή Φορτίου:**

   Η γραμμή φορτίου περιγράφεται από την εξίσωση:
   $
   I_C = \frac{V_{CC} - V_{CE}}{R_2}
   $

   Τα δύο χαρακτηριστικά σημεία της γραμμής φορτίου είναι:
   
   - Όταν $ V_{CE} = 0\,V$: 
     $
     I_C = I_{C,\text{max}} = \frac{V_{CC}}{R_2} = \frac{10}{1\cdot10^3} = 10\,mA
     $
     
   - Όταν $ I_C = 0\,A$: 
     $
     V_{CE} = V_{CC} = 10\,V
     $

---

#### **Σημείο Λειτουργίας Q**
Το σημείο λειτουργίας Q βρίσκεται στη διασταύρωση της γραμμής φορτίου με την χαρακτηριστική καμπύλη εξόδου του τρανζίστορ.

- Το σημείο Q είναι: 
$
Q(V_{CE}, I_C) = (7.85\,V, 2.15\,mA)
$

---

#### **Περιοχή Λειτουργίας**
Το τρανζίστορ λειτουργεί στην *ενεργή περιοχή* (active region), διότι:

- Η τάση $ V_{CE} > V_{BE} > 0.7\,V$.
- Υπάρχει σημαντικό $ I_C > 0\,A$, αλλά δεν έχει φτάσει στον κόρο.

Η θέση του σημείου Q πάνω στη χαρακτηριστική καμπύλη επιβεβαιώνει ότι το τρανζίστορ βρίσκεται στην ενεργή περιοχή.

---

#### **Διάγραμμα Γραμμής Φορτίου**
Στο συνημμένο διάγραμμα:

- Η μπλε γραμμή είναι η γραμμή φορτίου.
- Το κόκκινο σημείο αντιπροσωπεύει το σημείο λειτουργίας Q: $ (7.85\,V, 2.15\,mA) $.

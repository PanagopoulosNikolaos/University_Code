Παρακάτω δίνεται ένα ενδεικτικό παράδειγμα κώδικα σε Python (με τη βιβλιοθήκη matplotlib) που σχεδιάζει (α) τις θεωρητικές χαρακτηριστικές εξόδου του τρανζίστορ (Ic–VCE) για διάφορα ρεύματα βάσης και (β) την ευθεία φορτίου (load line), μαζί με το σημείο λειτουργίας Q. Η προσέγγιση αυτή βασίζεται στη θεωρία των διπολικών τρανζίστορ επαφής, όπου στη «δραστική περιοχή» ισχύει Ic ≈ β×Ib, ενώ στην περιοχή κορεσμού το ρεύμα συλλέκτη περιορίζεται λόγω της μικρής τάσης VCE. Σύμφωνα με την εισαγωγική βιβλιογραφία για τα τρανζίστορ μπορούμε να χρησιμοποιήσουμε την απλοποιημένη παραδοχή VCE(sat) ≈ 0,2 V. numpy as np

------------------------------------------------------------------------------

Επεξηγήσεις βασικών σημείων του κώδικα:  
• Για την προσέγγιση των χαρακτηριστικών εξόδου σχεδιάζουμε «οικογένειες» καμπυλών Ic–VCE για διάφορα ρεύματα βάσης Ib.  
• Σε κάθε καμπύλη, εάν η VCE βρίσκεται κάτω από την τάση κορεσμού (VCE_sat), τότε το τρανζίστορ θεωρείται κορεσμένο και το Ic «επιπεδώνεται». Για απλότητα, έχουμε Ic = β×Ib ακόμα και για VCE<VCE_sat, όμως συχνά αυτό το ρεύμα περιορίζεται περαιτέρω από το κύκλωμα.  
• Η ευθεία φορτίου (load line) προκύπτει από τη σχέση Ic = (VCC – VCE)/R2.  
• Το θεωρητικό σημείο λειτουργίας Q (Ib, IcQ, VCEQ) υπολογίζεται με βάση τις παραδοχές πόλωσης βάσης (R1, VBB και VBE=0,7 V). Στο διάγραμμα σχεδιάζεται ως «μαύρη τελεία» (ko).  

Από τη θέση του Q‐point (IcQ ≈ 2,15 mA και VCEQ ≈ 7,85 V) πάνω στο διάγραμμα, παρατηρούμε ότι το τρανζίστορ λειτουργεί στη «δραστική περιοχή» (forward‐active region), δηλαδή σε αρκετά υψηλή VCE ώστε να μη βρίσκεται σε κορεσμό και με ορατό καθορισμένο ρεύμα συλλέκτη από το ρεύμα βάσης.

````python
import numpy as np
import plotly.graph_objects as go

# -----------------------------------
# Παράμετροι τρανζίστορ και κυκλώματος
# -----------------------------------
beta = 500        # Συντελεστής ενίσχυσης
R2 = 1e3          # 1 kΩ
VCC = 10.0        # 10 V
VCE_sat = 0.2     # Τάση κορεσμού (προσεγγιστικά)
# Ρεύματα βάσης (διάφορες τιμές για τις καμπύλες)
Ib_values = [5e-6, 10e-6, 15e-6, 20e-6, 25e-6]

# -----------------------------------
# Προετοιμασία "άξονα" VCE
# -----------------------------------
VCE_points = np.linspace(0, 10, 200)  # από 0 έως 10 V

# Δημιουργία γραφήματος
fig = go.Figure()

# -----------------------------------
# Σχεδίαση χαρακτηριστικών εξόδου
# -----------------------------------
for Ib in Ib_values:
    Ic_curve = []
    for vce in VCE_points:
        if vce <= VCE_sat:
            Ic_curve.append(beta * Ib)
        else:
            Ic_curve.append(beta * Ib)
    fig.add_trace(go.Scatter(
        x=VCE_points, 
        y=Ic_curve, 
        mode='lines', 
        name=f'Ib = {Ib*1e6:.0f} μA',
        hoverinfo='text',
        hovertext=[f'VCE: {vce:.2f} V<br>IC: {ic:.6f} A' for vce, ic in zip(VCE_points, Ic_curve)]
    ))

# -----------------------------------
# Σχεδίαση ευθείας φορτίου (Load Line)
# -----------------------------------
Ic_load_line = (VCC - VCE_points) / R2
fig.add_trace(go.Scatter(
    x=VCE_points, 
    y=Ic_load_line, 
    mode='lines', 
    name='Load Line',
    line=dict(dash='dash'),
    hoverinfo='text',
    hovertext=[f'VCE: {vce:.2f} V<br>IC: {ic:.6f} A' for vce, ic in zip(VCE_points, Ic_load_line)]
))

# -----------------------------------
# Υπολογισμός σημείου λειτουργίας (Q-point)
# -----------------------------------
R1 = 1e6
VBB = 5.0
VBE = 0.7
Ib = (VBB - VBE) / R1  # ρεύμα βάσης
IcQ = beta * Ib        # ρεύμα συλλέκτη
VCEQ = VCC - IcQ * R2  # τάση συλλέκτη-εκπομπού

# Σχεδιάζουμε το σημείο Q
fig.add_trace(go.Scatter(
    x=[VCEQ], 
    y=[IcQ], 
    mode='markers+text', 
    name='Q-point',
    text=[f'Q\n(VCE={VCEQ:.2f} V,\nIC={IcQ*1e3:.2f} mA)'],
    textposition='top right',
    marker=dict(color='black', size=10)
))

# Προσθήκη τίτλων και ετικετών
fig.update_layout(
    title="Χαρακτηριστικές εξόδου τρανζίστορ με την ευθεία φορτίου",
    xaxis_title="VCE (V)",
    yaxis_title="IC (A)",
    xaxis=dict(showline=True, zeroline=True),
    yaxis=dict(showline=True, zeroline=True),
    legend=dict(x=1.05, y=1)  # Move the legend outside the main graph area
)

# Εμφάνιση γραφήματος
fig.show()
````

## Α) Υπολογισμός σημείου λειτουργίας Q

Για τον υπολογισμό του σημείου λειτουργίας Q ακολουθούμε τα εξής βήματα:

1. Υπολογισμός ρεύματος βάσης IB:
$$ I_B = \frac{V_{BB} - V_{BE}}{R_1} = \frac{5V - 0.7V}{1MΩ} = 4.3μΑ $$

2. Υπολογισμός ρεύματος συλλέκτη IC:
$$ I_C = β \cdot I_B = 500 \cdot 4.3μΑ = 2.15mA $$

3. Υπολογισμός τάσης VCE:
$$ V_{CE} = V_{CC} - I_C \cdot R_2 = 10V - 2.15mA \cdot 1kΩ = 7.85V $$

Επομένως, το σημείο λειτουργίας Q είναι:
$$ Q(V_{CE}, I_C) = (7.85V, 2.15mA) $$

## Β) Γραμμή φορτίου και περιοχή λειτουργίας

Η γραμμή φορτίου σχεδιάζεται με βάση δύο χαρακτηριστικά σημεία:
- Όταν VCE = 0V: IC = VCC/R2 = 10mA
- Όταν IC = 0A: VCE = VCC = 10V

Από τη θέση του σημείου Q στη χαρακτηριστική καμπύλη εξόδου (σχήμα 2), παρατηρούμε ότι το τρανζίστορ λειτουργεί στην **ενεργό περιοχή** καθώς:
- Η τάση VCE = 7.85V είναι μεγαλύτερη από την τάση κορεσμού VCEsat (~0.2V)
- Το σημείο Q βρίσκεται στη γραμμική περιοχή της χαρακτηριστικής καμπύλης εξόδου
